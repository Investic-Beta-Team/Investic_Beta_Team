```python
import pandas as pd
import numpy as np
import pandas_ta as ta
from tvDatafeed import TvDatafeed, Interval
import vectorbt as vbt
from tqdm import tqdm
import yfinance as yf
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/Users/euer/Documents/Study/Investment_Project/2.Longrun/EURUSD_60M.csv')
df = df.set_index('time')
```

```python
lst = []
for i in range(5,50):
    for j in range(i+1,51):
        for k in range(60,110,20):
            lst.append([i,j,k])
```

```python
for i in range(5,101):
    df[f'ema{i}'] = df.ta.ema(i)
```

```python
for start , end in tqdm(zip(range(0,len(df),720),range(1440,len(df),720))):
    tmp = df.iloc[start:end]
    entries = tmp[['close']].drop('close',axis=1)
    exits = tmp[['close']].drop('close',axis=1)
    for i in range(len(lst)):
        fast = lst[i][0]
        slow = lst[i][1]
        filter = lst[i][2]
        tmp['signal'] = (tmp.ta.long_run(tmp[f'ema{fast}'],tmp[f'ema{slow}'],length=10)) & (tmp['close'] > tmp[f'ema{filter}'])
        signal_vbt = tmp.ta.tsignals(tmp.signal,asbool=True,append=True)
        signal_vbt.rename(columns={'TS_Entries':f'fast_{fast}_slow{slow}_filter{filter}_entry','TS_Exits':f'fast_{fast}_slow{slow}_filter{filter}_exit'},inplace=True)
        entries = entries.join(signal_vbt[f'fast_{fast}_slow{slow}_filter{filter}_entry'])
        exits = exits.join(signal_vbt[f'fast_{fast}_slow{slow}_filter{filter}_exit'])
    globals()[f'entries_{start}'] = entries
    globals()[f'exits_{start}'] = exits
```
